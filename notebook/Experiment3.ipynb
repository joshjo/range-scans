```python
import pandas as pd
import matplotlib.pyplot as plt

# Updated to July 25th 07:11 GMT-5

def profiling(url, groupby=['num queries', 'strategy']):
    exp1df = pd.read_csv(url)
    exp1df_group = exp1df[exp1df.iter > 1].groupby(groupby).mean()
    return exp1df_group
```

```python
df3a = profiling("https://raw.githubusercontent.com/joshjo/range-scans/master/profiling/2020-08-11/experiment_3a.csv", groupby=['leaf size', 'strategy'])
df3a
```
